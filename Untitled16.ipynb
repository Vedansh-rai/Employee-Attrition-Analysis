```python
# Install required packages
!pip install kagglehub pandas numpy matplotlib seaborn scikit-learn openpyxl plotly -q

import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Download the IBM HR Analytics dataset
print("Downloading IBM HR Analytics Attrition Dataset...")
path = kagglehub.dataset_download("pavansubhasht/ibm-hr-analytics-attrition-dataset")
print(f"Dataset downloaded to: {path}")

# List files in the downloaded directory
import os
print("\nFiles in the dataset directory:")
for file in os.listdir(path):
    print(f"  ‚Ä¢ {file}")
```

```python
# Load the dataset (usually named WA_Fn-UseC_-HR-Employee-Attrition.csv)
dataset_file = os.path.join(path, "WA_Fn-UseC_-HR-Employee-Attrition.csv")
df = pd.read_csv(dataset_file)

print("="*80)
print("IBM HR ANALYTICS ATTRITION DATASET - COMPLETE ANALYSIS")
print("="*80)

print(f"\nüìä Dataset Shape: {df.shape}")
print(f"üìù Columns: {len(df.columns)}")
print("\nFirst 5 rows:")
display(df.head())
print("\nDataset Info:")
df.info()
print("Dataset Columns:")
for i, col in enumerate(df.columns, 1):
    print(f"{i:2}. {col}")
```

```python
# ==============================================
# 1. DATA EXPLORATION AND CLEANING
# ==============================================

print("="*80)
print("1. DATA EXPLORATION AND CLEANING")
print("="*80)

# Check for missing values
print("\nüîç Missing Values Analysis:")
missing_values = df.isnull().sum()
missing_pct = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percentage': missing_pct
})
print(missing_df[missing_df['Missing_Count'] > 0])

# If no missing values (as expected with this dataset)
if missing_df['Missing_Count'].sum() == 0:
    print("‚úÖ No missing values found in the dataset!")

# Check data types
print("\nüìä Data Types:")
print(df.dtypes.value_counts())

# Identify categorical and numerical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print(f"\nüìã Categorical columns ({len(categorical_cols)}):")
for col in categorical_cols:
    print(f"  ‚Ä¢ {col}: {df[col].nunique()} unique values")

print(f"\nüìà Numerical columns ({len(numerical_cols)}):")
print(", ".join(numerical_cols[:10]))
if len(numerical_cols) > 10:
    print(f"... and {len(numerical_cols)-10} more")

# Create binary attrition flag
df['Attrition_Flag'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# Save cleaned dataset
df.to_csv('/content/IBM_HR_Data_Cleaned.csv', index=False)
print("\n‚úÖ Cleaned dataset saved as 'IBM_HR_Data_Cleaned.csv'")
```

```python
# ==============================================
# 2. ATTRITION OVERVIEW
# ==============================================

print("="*80)
print("2. ATTRITION OVERVIEW")
print("="*80)

# Attrition distribution
attrition_counts = df['Attrition'].value_counts()
attrition_pct = df['Attrition'].value_counts(normalize=True) * 100

print(f"\nüìä Attrition Distribution:")
print(f"‚Ä¢ Yes (Attrited): {attrition_counts['Yes']} employees ({attrition_pct['Yes']:.1f}%)")
print(f"‚Ä¢ No (Retained): {attrition_counts['No']} employees ({attrition_pct['No']:.1f}%)")

# Create separate dataframes for attrited and retained employees
attrited = df[df['Attrition'] == 'Yes']
retained = df[df['Attrition'] == 'No']

print(f"\nüìù Dataset Summary:")
print(f"Total Employees: {len(df)}")
print(f"Attrited Employees: {len(attrited)}")
print(f"Retained Employees: {len(retained)}")
```

```python
# ==============================================
# 3. COMPREHENSIVE VISUALIZATIONS
# ==============================================

print("="*80)
print("3. CREATING COMPREHENSIVE VISUALIZATIONS")
print("="*80)

# Create a directory for saving plots
!mkdir -p /content/attrition_plots

# Set up figure for multiple plots
fig = plt.figure(figsize=(20, 16))

# 1. Attrition Distribution (Pie Chart)
ax1 = plt.subplot(3, 4, 1)
colors = ['#4CAF50', '#F44336']
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%',
        colors=colors, startangle=90, explode=(0.1, 0))
plt.title('Overall Attrition Distribution', fontsize=14, fontweight='bold')
plt.axis('equal')

# 2. Age Distribution Comparison
ax2 = plt.subplot(3, 4, 2)
sns.histplot(data=attrited, x='Age', color='red', alpha=0.5, label='Attrited', kde=True, bins=20)
sns.histplot(data=retained, x='Age', color='blue', alpha=0.5, label='Retained', kde=True, bins=20)
plt.title('Age Distribution: Attrited vs Retained', fontsize=14, fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()

# 3. Monthly Income Distribution
ax3 = plt.subplot(3, 4, 3)
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette=colors)
plt.title('Monthly Income by Attrition Status', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income ($)')

# 4. Job Satisfaction
ax4 = plt.subplot(3, 4, 4)
satisfaction_counts = pd.crosstab(df['Attrition'], df['JobSatisfaction'])
satisfaction_counts.plot(kind='bar', ax=ax4)
plt.title('Job Satisfaction by Attrition Status', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.legend(title='Satisfaction Level', bbox_to_anchor=(1.05, 1), loc='upper left')

# 5. Years at Company
ax5 = plt.subplot(3, 4, 5)
sns.kdeplot(data=attrited['YearsAtCompany'], label='Attrited', color='red', fill=True)
sns.kdeplot(data=retained['YearsAtCompany'], label='Retained', color='blue', fill=True)
plt.title('Tenure Distribution: Attrited vs Retained', fontsize=14, fontweight='bold')
plt.xlabel('Years at Company')
plt.ylabel('Density')
plt.legend()

# 6. Overtime Impact
ax6 = plt.subplot(3, 4, 6)
overtime_attrition = pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100
overtime_attrition.plot(kind='bar', ax=ax6)
plt.title('Attrition Rate by Overtime Status', fontsize=14, fontweight='bold')
plt.xlabel('Overtime')
plt.ylabel('Percentage (%)')
plt.legend(title='Attrition')

# 7. Department-wise Attrition
ax7 = plt.subplot(3, 4, 7)
dept_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack() * 100
dept_attrition['Yes'].plot(kind='bar', color='coral')
plt.title('Attrition Rate by Department', fontsize=14, fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45)

# 8. Work Life Balance
ax8 = plt.subplot(3, 4, 8)
wlb_attrition = pd.crosstab(df['WorkLifeBalance'], df['Attrition'], normalize='index') * 100
wlb_attrition.plot(kind='bar', ax=ax8)
plt.title('Attrition Rate by Work Life Balance', fontsize=14, fontweight='bold')
plt.xlabel('Work Life Balance Score')
plt.ylabel('Percentage (%)')
plt.legend(title='Attrition')

# 9. Stock Option Level
ax9 = plt.subplot(3, 4, 9)
stock_attrition = pd.crosstab(df['StockOptionLevel'], df['Attrition'], normalize='index') * 100
stock_attrition.plot(kind='bar', ax=ax9)
plt.title('Attrition Rate by Stock Option Level', fontsize=14, fontweight='bold')
plt.xlabel('Stock Option Level')
plt.ylabel('Percentage (%)')
plt.legend(title='Attrition')

# 10. Job Role Attrition
ax10 = plt.subplot(3, 4, 10)
jobrole_counts = df['JobRole'].value_counts()
top_roles = jobrole_counts.head(6).index
top_roles_data = df[df['JobRole'].isin(top_roles)]
role_attrition = top_roles_data.groupby('JobRole')['Attrition'].value_counts(normalize=True).unstack() * 100
role_attrition['Yes'].plot(kind='bar', color='purple')
plt.title('Attrition Rate by Job Role (Top 6)', fontsize=14, fontweight='bold')
plt.xlabel('Job Role')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45)

# 11. Business Travel
ax11 = plt.subplot(3, 4, 11)
travel_attrition = pd.crosstab(df['BusinessTravel'], df['Attrition'], normalize='index') * 100
travel_attrition.plot(kind='bar', ax=ax11)
plt.title('Attrition Rate by Business Travel', fontsize=14, fontweight='bold')
plt.xlabel('Business Travel')
plt.ylabel('Percentage (%)')
plt.legend(title='Attrition')
plt.xticks(rotation=45)

# 12. Marital Status
ax12 = plt.subplot(3, 4, 12)
marital_attrition = pd.crosstab(df['MaritalStatus'], df['Attrition'], normalize='index') * 100
marital_attrition.plot(kind='bar', ax=ax12)
plt.title('Attrition Rate by Marital Status', fontsize=14, fontweight='bold')
plt.xlabel('Marital Status')
plt.ylabel('Percentage (%)')
plt.legend(title='Attrition')

plt.tight_layout()
plt.savefig('/content/attrition_plots/comprehensive_dashboard.png', dpi=300, bbox_inches='tight')
print("‚úÖ Comprehensive dashboard saved as 'attrition_plots/comprehensive_dashboard.png'")
plt.show()
```

```python
# ==============================================
# 4. STATISTICAL ANALYSIS
# ==============================================

print("="*80)
print("4. STATISTICAL ANALYSIS")
print("="*80)

# Key metrics comparison
print("\nüìä KEY METRICS COMPARISON:")
print("-"*50)

metrics = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction',
           'WorkLifeBalance', 'YearsSinceLastPromotion', 'TotalWorkingYears',
           'DistanceFromHome', 'PercentSalaryHike']

summary_data = []
for metric in metrics:
    att_mean = attrited[metric].mean()
    ret_mean = retained[metric].mean()
    diff = att_mean - ret_mean
    pct_diff = (diff / ret_mean) * 100

    summary_data.append({
        'Metric': metric,
        'Attrited': f"{att_mean:.2f}",
        'Retained': f"{ret_mean:.2f}",
        'Difference': f"{diff:+.2f}",
        'Pct_Diff': f"{pct_diff:+.1f}%"
    })

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

# Save summary statistics
summary_df.to_csv('/content/attrition_summary_statistics.csv', index=False)
print("\n‚úÖ Summary statistics saved as 'attrition_summary_statistics.csv'")
```

```python
# ==============================================
# 5. CORRELATION ANALYSIS
# ==============================================

print("="*80)
print("5. CORRELATION ANALYSIS")
print("="*80)

# Select numerical columns for correlation
numerical_cols_for_corr = [col for col in numerical_cols if col not in ['EmployeeCount', 'EmployeeNumber', 'StandardHours']]

# Calculate correlations with attrition
correlations = df[numerical_cols_for_corr + ['Attrition_Flag']].corr()['Attrition_Flag'].sort_values(ascending=False)
correlations = correlations.drop('Attrition_Flag')

print("\nüîó TOP CORRELATIONS WITH ATTRITION:")
print("-"*50)
print("Positive Correlations (Higher value = More likely to leave):")
for idx, (feature, corr) in enumerate(correlations[correlations > 0].head(10).items(), 1):
    print(f"{idx:2}. {feature:30}: {corr:+.3f}")

print("\nNegative Correlations (Higher value = Less likely to leave):")
for idx, (feature, corr) in enumerate(correlations[correlations < 0].head(10).items(), 1):
    print(f"{idx:2}. {feature:30}: {corr:+.3f}")

# Save correlations
pd.DataFrame(correlations, columns=['Correlation']).to_csv('/content/attrition_correlations.csv')
print("\n‚úÖ Correlation analysis saved as 'attrition_correlations.csv'")

# Correlation heatmap for top 15 features
plt.figure(figsize=(12, 10))
top_features = correlations.abs().sort_values(ascending=False).head(15).index.tolist()
corr_matrix = df[top_features + ['Attrition_Flag']].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap: Top 15 Features with Attrition', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/attrition_plots/correlation_heatmap.png', dpi=300, bbox_inches='tight')
print("‚úÖ Correlation heatmap saved")
plt.show()
```

```python
# ==============================================
# 6. DECISION TREE ANALYSIS - TOP PREDICTORS
# ==============================================

print("="*80)
print("6. DECISION TREE ANALYSIS - TOP PREDICTORS")
print("="*80)

# Prepare data for modeling
# Exclude non-predictive columns
exclude_cols = ['EmployeeCount', 'EmployeeNumber', 'StandardHours',
                'Over18', 'Attrition', 'Attrition_Flag']

feature_cols = [col for col in df.columns if col not in exclude_cols]

# Prepare feature matrix
X = df[feature_cols].copy()
y = df['Attrition_Flag']

# Convert categorical variables to dummy variables
categorical_cols = X.select_dtypes(include=['object']).columns
print(f"\nConverting {len(categorical_cols)} categorical columns to dummy variables...")

if len(categorical_cols) > 0:
    X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

print(f"Final feature matrix shape: {X.shape}")

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nüìä Data Split:")
print(f"Training set: {X_train.shape[0]} samples")
print(f"Testing set: {X_test.shape[0]} samples")
print(f"Attrition rate in training: {y_train.mean()*100:.1f}%")
print(f"Attrition rate in testing: {y_test.mean()*100:.1f}%")

# Build decision tree
print("\nüå≥ Building Decision Tree Model...")
dt_model = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42,
    class_weight='balanced'
)

dt_model.fit(X_train, y_train)

# Evaluate model
y_pred = dt_model.predict(X_test)

print("\nüìà MODEL EVALUATION:")
print("-"*40)
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nüéØ CONFUSION MATRIX:")
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm,
                     index=['Actual Retained', 'Actual Attrited'],
                     columns=['Predicted Retained', 'Predicted Attrited'])
print(cm_df)

# Feature importance
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': dt_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nüèÜ TOP 15 ATTRITION PREDICTORS:")
print("-"*50)
for idx, row in feature_importance.head(15).iterrows():
    print(f"{row['Importance']*100:6.2f}% - {row['Feature']}")

# Save feature importance
feature_importance.to_csv('/content/attrition_feature_importance.csv', index=False)
print("\n‚úÖ Feature importance saved as 'attrition_feature_importance.csv'")
```

```python
# Visualize decision tree (top 3 levels for clarity)
plt.figure(figsize=(20, 12))
tree.plot_tree(dt_model,
               feature_names=X.columns,
               class_names=['Retained', 'Attrited'],
               filled=True,
               rounded=True,
               max_depth=3,
               fontsize=10,
               proportion=True,
               impurity=False)
plt.title('Decision Tree for Attrition Prediction (Top 3 Levels)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/attrition_plots/decision_tree.png', dpi=300, bbox_inches='tight')
print("‚úÖ Decision tree visualization saved")
plt.show()

# Feature importance bar chart
plt.figure(figsize=(12, 8))
top_features = feature_importance.head(10)
sns.barplot(data=top_features, x='Importance', y='Feature', palette='viridis')
plt.title('Top 10 Attrition Predictors (Feature Importance)', fontsize=16, fontweight='bold')
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig('/content/attrition_plots/feature_importance.png', dpi=300, bbox_inches='tight')
print("‚úÖ Feature importance chart saved")
plt.show()
```

```python
# ==============================================
# 7. HIGH-RISK EMPLOYEE IDENTIFICATION
# ==============================================

print("="*80)
print("7. HIGH-RISK EMPLOYEE IDENTIFICATION")
print("="*80)

# Use decision tree predictions to identify high-risk employees
df['Predicted_Attrition_Probability'] = dt_model.predict_proba(X)[:, 1]

# Identify high-risk employees (top 20% probability)
high_risk_threshold = df['Predicted_Attrition_Probability'].quantile(0.8)
high_risk_employees = df[df['Predicted_Attrition_Probability'] >= high_risk_threshold]

print(f"\nüî¥ HIGH-RISK EMPLOYEE PROFILE:")
print("-"*50)
print(f"Number of high-risk employees: {len(high_risk_employees)} ({len(high_risk_employees)/len(df)*100:.1f}%)")
print(f"Risk threshold (probability): {high_risk_threshold:.3f}")

# Characteristics of high-risk employees
print(f"\nüìä Characteristics of High-Risk Employees:")

# Department distribution
if 'Department' in df.columns:
    dept_dist = high_risk_employees['Department'].value_counts(normalize=True) * 100
    print(f"\nüìã Department Distribution:")
    for dept, pct in dept_dist.items():
        print(f"  ‚Ä¢ {dept}: {pct:.1f}%")

# Job Role distribution
if 'JobRole' in df.columns:
    role_dist = high_risk_employees['JobRole'].value_counts(normalize=True).head(5) * 100
    print(f"\nüë• Top 5 Job Roles:")
    for role, pct in role_dist.items():
        print(f"  ‚Ä¢ {role}: {pct:.1f}%")

# Overtime
if 'OverTime' in df.columns:
    ot_pct = (high_risk_employees['OverTime'] == 'Yes').mean() * 100
    print(f"\n‚è∞ Overtime: {ot_pct:.1f}% work overtime")

# Average metrics
print(f"\nüìà Average Metrics for High-Risk Employees:")
print(f"  ‚Ä¢ Monthly Income: ${high_risk_employees['MonthlyIncome'].mean():,.0f}")
print(f"  ‚Ä¢ Age: {high_risk_employees['Age'].mean():.1f} years")
print(f"  ‚Ä¢ Years at Company: {high_risk_employees['YearsAtCompany'].mean():.1f} years")
print(f"  ‚Ä¢ Job Satisfaction: {high_risk_employees['JobSatisfaction'].mean():.2f}/4")
print(f"  ‚Ä¢ Work Life Balance: {high_risk_employees['WorkLifeBalance'].mean():.2f}/4")

# Save high-risk employees
high_risk_cols = ['Age', 'Department', 'JobRole', 'MonthlyIncome', 'YearsAtCompany',
                  'JobSatisfaction', 'WorkLifeBalance', 'OverTime', 'BusinessTravel',
                  'MaritalStatus', 'Predicted_Attrition_Probability', 'Attrition']

high_risk_cols = [col for col in high_risk_cols if col in df.columns]
high_risk_employees[high_risk_cols].sort_values('Predicted_Attrition_Probability', ascending=False).to_csv(
    '/content/high_risk_employees.csv', index=False)

print(f"\n‚úÖ High-risk employees saved as 'high_risk_employees.csv'")
```

```python
# ==============================================
# 8. EXCEL DASHBOARD CREATION
# ==============================================

print("="*80)
print("8. CREATING EXCEL DASHBOARD")
print("="*80)

try:
    with pd.ExcelWriter('/content/IBM_Attrition_Dashboard.xlsx', engine='openpyxl') as writer:
        # 1. Overall Summary
        summary_data = pd.DataFrame({
            'Metric': ['Total Employees', 'Attrited Employees', 'Retained Employees',
                      'Attrition Rate', 'High-Risk Employees', 'Average Age',
                      'Average Monthly Income', 'Average Tenure (Years)'],
            'Value': [
                len(df),
                len(attrited),
                len(retained),
                f"{attrition_pct['Yes']:.1f}%",
                len(high_risk_employees),
                f"{df['Age'].mean():.1f}",
                f"${df['MonthlyIncome'].mean():,.0f}",
                f"{df['YearsAtCompany'].mean():.1f}"
            ]
        })
        summary_data.to_excel(writer, sheet_name='Overall_Summary', index=False)

        # 2. Department Analysis
        dept_analysis = df.groupby('Department').agg(
            Total_Employees=('Department', 'count'),
            Attrited=('Attrition_Flag', 'sum'),
            Attrition_Rate=('Attrition_Flag', lambda x: x.mean() * 100),
            Avg_Monthly_Income=('MonthlyIncome', 'mean'),
            Avg_Age=('Age', 'mean'),
            Avg_Job_Satisfaction=('JobSatisfaction', 'mean')
        ).round(2)
        dept_analysis.to_excel(writer, sheet_name='Department_Analysis')

        # 3. Job Role Analysis
        jobrole_analysis = df.groupby('JobRole').agg(
            Total_Employees=('JobRole', 'count'),
            Attrited=('Attrition_Flag', 'sum'),
            Attrition_Rate=('Attrition_Flag', lambda x: x.mean() * 100),
            Avg_Monthly_Income=('MonthlyIncome', 'mean')
        ).round(2).sort_values('Attrition_Rate', ascending=False)
        jobrole_analysis.to_excel(writer, sheet_name='JobRole_Analysis')

        # 4. Overtime Analysis
        overtime_analysis = pd.crosstab(df['OverTime'], df['Attrition'],
                                       margins=True, margins_name='Total')
        overtime_analysis['Attrition_Rate'] = (overtime_analysis['Yes'] / overtime_analysis['Total']) * 100
        overtime_analysis.to_excel(writer, sheet_name='Overtime_Analysis')

        # 5. Marital Status Analysis
        marital_analysis = pd.crosstab(df['MaritalStatus'], df['Attrition'],
                                      margins=True, margins_name='Total')
        marital_analysis['Attrition_Rate'] = (marital_analysis['Yes'] / marital_analysis['Total']) * 100
        marital_analysis.to_excel(writer, sheet_name='MaritalStatus_Analysis')

        # 6. Business Travel Analysis
        travel_analysis = pd.crosstab(df['BusinessTravel'], df['Attrition'],
                                     margins=True, margins_name='Total')
        travel_analysis['Attrition_Rate'] = (travel_analysis['Yes'] / travel_analysis['Total']) * 100
        travel_analysis.to_excel(writer, sheet_name='BusinessTravel_Analysis')

        # 7. Feature Importance
        feature_importance.head(20).to_excel(writer, sheet_name='Feature_Importance', index=False)

        # 8. High-Risk Employees
        high_risk_employees[high_risk_cols].sort_values('Predicted_Attrition_Probability', ascending=False).head(50).to_excel(
            writer, sheet_name='High_Risk_Employees', index=False)

        # 9. Correlation Matrix
        corr_matrix_df = pd.DataFrame(corr_matrix)
        corr_matrix_df.to_excel(writer, sheet_name='Correlation_Matrix')

    print("‚úÖ Excel dashboard saved as 'IBM_Attrition_Dashboard.xlsx'")

except Exception as e:
    print(f"‚ö†Ô∏è Error creating Excel dashboard: {str(e)}")
```
